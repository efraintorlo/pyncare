```python
def compact_dyn_sys_phi2(init, t=None, model_pars=[]):
        if t is None:
            t = 1.0  # This is used to allows a call from "plot_vector_flow()"
        X = init[0]
        Y = init[1]
        # the model equations
        Z = np.sqrt(1.0 - X**2.0 - Y**2.0)
        A = 3.0 * Y * np.sqrt(X**2.0 + Y**2.0) / Z
        X_dot = Y + X * Y * A
        Y_dot = -X + (Y**2.0 - 1.0) * A
        return [X_dot, Y_dot]

def compact_dyn_sys_phi2_phi4(init, t=None, model_pars=[]):
    if t is None:
        t = 1.0  # This is used to allows a call from "plot_vector_flow()"
    X = init[0]
    Y = init[1]
    # alpha_sqd = 6.0E-6
    # alpha_sqd = 6.0
    alpha_sqd = 10.0
    # the model equations
    Z = np.sqrt(1.0 - X**2.0 - Y**2.0)
    A = 3.0 * Y * np.sqrt(X**2.0 + Y**2.0 + 3.0 * alpha_sqd * (X**4.0 / Z**2.0)) / Z
    X_dot = Y + X * Y * A + alpha_sqd * X**4.0 * Y / Z**2 
    Y_dot = -X + (Y**2.0 - 1.0) * A + alpha_sqd * X**3.0 * (Y**2.0 - 1.0) / Z**2 
    return [X_dot, Y_dot]
```

```python
dynsys = PoincareCompact(model=compact_dyn_sys_phi2_phi4,
                         model_pars=[],
                         var_names=var_names,
                         Ndim=2,
                         orbits=orbits,
                         lines=None,
                         #colors='bright',
                         colors='black',
                         )
```
