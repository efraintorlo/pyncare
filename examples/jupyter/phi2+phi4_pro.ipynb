```python
def dyn_sys_phi2(init, t=None, modelpars=[]):
        if t is None:
            t = 1.0  # This is used to allows a call from "plot_vector_flow()"
        x = init[0]
        y = init[1]
        # the model equations
        z = np.sqrt(x**2.0 + y**2.0 )
        x_dot = y
        y_dot = - x - 3.0*y*z 
        return [x_dot, y_dot]
    
def dyn_sys_phi2_phi4(init, t=None, modelpars=[]):
        if t is None:
            t = 1.0  # This is used to allows a call from "plot_vector_flow()"
        x = init[0]
        y = init[1]
        alpha_sqd = 70.0
        # the model equations
        z = np.sqrt(x**2.0 + y**2.0 + 3.0*alpha_sqd*x**4.0)
        x_dot = y
        y_dot = - x - 3.0*y*z - alpha_sqd*x**3.0 
        return [x_dot, y_dot]
```
